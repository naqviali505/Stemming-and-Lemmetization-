###1.1 Write a function that reurns Count number of lines in txt file

```python
def NumOfLines(filename):  
  #Write your logic here
  with open(filename) as fp:
    count = sum(1 for line in fp)
  return count 
```

```python
count = NumOfLines("text_file.txt")
print("Number of Lines are ", count)
```

###1.2 Write a function that reurns Count number of words in txt file

```python
def NumOfWords(filename):
  #Write your logic here
  with open(filename) as fp:
    data = fp.read()
    words = len(data.split())
  return words
```

```python
num = NumOfWords("text_file.txt")
print("Number of words are ", num)
```

##Task 2
Write a function that reurns a dataframe with two columns having stop words in one col and their count in the other.

```python
import nltk
nltk.download('stopwords')
```

```python
from pandas import DataFrame
from nltk.corpus import stopwords
from nltk import regexp_tokenize
def CreateDF(fileName):
  #Write your logic here
  with open(fileName) as fp:
    data = fp.read()
    data = regexp_tokenize(data,pattern=r"\w+")
    stop_words = stopwords.words('english')
    wordcount=dict()
    for word in data:
      if word in stop_words:
        if word in wordcount:
          wordcount[word]+=1
        else:
          wordcount[word]=1
    df = DataFrame(list(wordcount.items()))
    return df
```

```python
df = CreateDF("text_file.txt")
df
```

##Task 3
Now generate a txt file named RollnumberWS.txt having all content of the given file but excluding stopwords.

```python
from nltk.corpus import stopwords
from nltk import regexp_tokenize
def WithoutSW(filename, filetowritein):
  #Write your logic here
  with open(filename) as fp:
    data = fp.read()
    data = regexp_tokenize(data,pattern=r"\w+")
    stop_words = stopwords.words('english')
    for word in data:
      if not word in stop_words:
        fpp = open(filetowritein,"a")
        fpp.write(word+" ")
        fpp.close()
```

```python
WithoutSW("text_file.txt", "L191015WS.txt")
```

##Task 4
Write a function that displays all words having length greater than 5.

```python
from nltk import regexp_tokenize
def GreaterThan5(fileName):
  # Write your logic here
  with open(fileName) as fp:
    data = fp.read()
    data = regexp_tokenize(data, pattern=r'\w+')
    for x in data:
      if(len(x) > 5):
        print(x)
```

##Task 5
Write a function that returns the count of all words which end at y.

```python
from nltk import regexp_tokenize
def EndAty(filename):
  # Write your logic here
  yEnding = []
  with open(filename) as fp:
    data = fp.read()
    data = regexp_tokenize(data, pattern=r'\w+')
    for x in data:
      if(x[-1] == 'y'):
        yEnding.append(x)
  return len(yEnding)
```

```python
w = EndAty("text_file.txt")
w
```

##Task 6
Write a function that prints the count of lowercase letters in the file.

```python
def CountLower(filename):
  #Write your logic here
  with open(filename) as fp:
    data = fp.read()
    count = 0
    for x in range(len(data)):
      if(data[x].islower()):
        count += 1
    print(count)
```

```python
CountLower("text_file.txt")
```

##Task 7
Write a function ZICount() that prints the count of all occurences of Z and I (including small cases z and i too)
For example: Hi! I am Zenia. It was so good to see you! 
####Output: 
I or i : 4

Z or z : 1


```python
def ZICount(filename):
  #Write your logic here
  with open(filename) as fp:
    data = fp.read()
    icount=zcount=0
    for i in data:
      if(i.lower()=='i'):
        icount+=1
      elif(i.lower()=='z'):
        zcount+=1
    print("I or i:",icount)
    print("Z or z:",zcount)
```

```python
ZICount("text_file.txt")
```

##Task 8
Write a function to display the content of a file in descending order (based on the word count)

Content: I am confused to start it but it won't be done until I give it a try

Output: I it to a am be confused done give start try until wont.


```python
import operator
from nltk import regexp_tokenize
def CountAsc(filename):
  #Write your logic here
  with open(filename) as fp:
    data = fp.read()
    data = regexp_tokenize(data,pattern=r"\w+")
    wordcount=dict()
    for word in data:
      if word in wordcount:
        wordcount[word]+=1
      else:
        wordcount[word]=1
    wordcount = dict(sorted(wordcount.items(),key=operator.itemgetter(1),reverse=True))
    print(wordcount.keys())
```

```python
CountAsc("text_file.txt")
```

##Task 9
Perform lemmatisation on the given file and generate a txt file named Rollnumber_lemmatized.txt having all content of the given file.




```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
#Write your logic here
from nltk import regexp_tokenize
from nltk import WordNetLemmatizer
def lemmatization(filename):
  with open(filename) as fp:
    data = fp.read()
    data = regexp_tokenize(data,pattern=r"\w+")
    lemmatizer = WordNetLemmatizer()
    for word in data:
      root = lemmatizer.lemmatize(word)
      fpp = open("L191015_lemmatized.txt","a")
      fpp.write(root+" ")
      fpp.close()
```

```python
lemmatization("text_file.txt")
```

##Task 10
Perform stemming on the given file and generate a txt file named Rollnumber_stemming.txt having all content of the given file.


```python
from nltk import regexp_tokenize
from nltk import PorterStemmer
def stemming(filename):
  with open(filename) as fp:
    data = fp.read()
    data = regexp_tokenize(data,pattern=r"\w+")
    ps = PorterStemmer()
    for word in data:
      root = ps.stem(word)
      fpp = open("L191015_stemming.txt","a")
      fpp.write(root+" ")
      fpp.close()
```

```python
#Write your logic here
stemming("text_file.txt")
```
